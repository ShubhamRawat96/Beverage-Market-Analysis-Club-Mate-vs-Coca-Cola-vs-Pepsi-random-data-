```python
import pandas as pd

# Update the path if your file is in a different folder
file_path = r"D:\club-mate\data\processed\club_mate_clean.csv"

# Load CSV
df = pd.read_csv(file_path)

# Preview first 5 rows
df.head()
```

```python
# Shape of the dataset
print("Shape (rows, cols):", df.shape)

# Missing values per column
print("\nMissing values:\n", df.isna().sum())

# Duplicate rows
print("\nDuplicate rows:", df.duplicated().sum())

# Data types
print("\nData types:\n", df.dtypes)

# Convert 'Date' to datetime if not already
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Total Revenue, Cost, Profit
total_revenue = df['Revenue'].sum()
total_cost = df['Total_Cost'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_revenue) * 100

print(f"\nTotal Revenue: €{total_revenue:,.2f}")
print(f"Total Cost: €{total_cost:,.2f}")
print(f"Total Profit: €{total_profit:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")

# Top 5 Products by Revenue
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(5)
print("\nTop Products by Revenue:\n", top_products)

# Top 5 Regions by Revenue
top_regions = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False).head(5)
print("\nTop Regions by Revenue:\n", top_regions)
```

```python
import matplotlib.pyplot as plt

# Ensure Date is datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Create Year-Month column
df['YearMonth'] = df['Date'].dt.to_period('M')

# --- Revenue & Profit trend by Month ---
monthly_trend = df.groupby('YearMonth').agg({
    'Revenue':'sum',
    'Profit':'sum'
}).reset_index()

# Convert YearMonth back to string for plotting
monthly_trend['YearMonth'] = monthly_trend['YearMonth'].astype(str)

plt.figure(figsize=(12,6))
plt.plot(monthly_trend['YearMonth'], monthly_trend['Revenue'], marker='o', label='Revenue (€)')
plt.plot(monthly_trend['YearMonth'], monthly_trend['Profit'], marker='s', label='Profit (€)')
plt.xticks(rotation=45)
plt.title("Monthly Revenue & Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount (€)")
plt.legend()
plt.tight_layout()
plt.show()

# --- Club-Mate vs Coca-Cola vs Pepsi Revenue over Time ---
brand_trend = df.groupby(['YearMonth','Product'])['Revenue'].sum().reset_index()
brand_pivot = brand_trend.pivot(index='YearMonth', columns='Product', values='Revenue').fillna(0)

brand_pivot.plot(figsize=(12,6), marker='o')
plt.title("Monthly Revenue Trend by Product")
plt.xlabel("Month")
plt.ylabel("Revenue (€)")
plt.xticks(rotation=45)
plt.legend(title="Product")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- Prepare Monthly Market Share Data ---
brand_monthly = (
    df.groupby([df["Date"].dt.to_period("M"), "Product"])["Revenue"]
    .sum()
    .reset_index()
)

brand_monthly["Date"] = brand_monthly["Date"].astype(str)

# Pivot for market share
brand_pivot = brand_monthly.pivot(index="Date", columns="Product", values="Revenue").fillna(0)

# Calculate % share
brand_share = brand_pivot.div(brand_pivot.sum(axis=1), axis=0) * 100

# --- 1. Stacked Area Chart (Market Share %) ---
plt.figure(figsize=(12,6))
plt.stackplot(
    brand_share.index,
    brand_share["Club-Mate"],
    brand_share["Coca-Cola"],
    brand_share["Pepsi"],
    labels=["Club-Mate", "Coca-Cola", "Pepsi"],
    alpha=0.8
)
plt.title("Monthly Market Share (%) - Club-Mate vs Coca-Cola vs Pepsi")
plt.xlabel("Month")
plt.ylabel("Market Share (%)")
plt.legend(loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(save_folder, "market_share_stacked_area.png"))
plt.show()

# --- 2. Multi-Line Chart (Market Share %) ---
plt.figure(figsize=(12,6))
for brand in ["Club-Mate", "Coca-Cola", "Pepsi"]:
    plt.plot(brand_share.index, brand_share[brand], marker="o", label=brand)

plt.title("Monthly Market Share Trend (%)")
plt.xlabel("Month")
plt.ylabel("Market Share (%)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(save_folder, "market_share_trend_lines.png"))
plt.show()
```

```python
import pandas as pd
import os

# Path to the existing summary file
excel_path = r"D:\club-mate\reports\clubmate_summary.xlsx"

# --- Prepare Dashboard KPIs ---
dashboard = pd.DataFrame({
    "Metric": [
        "Total Revenue (€)",
        "Total Cost (€)",
        "Total Profit (€)",
        "Profit Margin (%)",
        "Top Product by Revenue",
        "Top Region by Revenue",
        "Top Channel by Revenue"
    ],
    "Value": [
        round(df["Revenue"].sum(), 2),
        round(df["Total_Cost"].sum(), 2),
        round(df["Profit"].sum(), 2),
        round((df["Profit"].sum() / df["Revenue"].sum()) * 100, 2),
        df.groupby("Product")["Revenue"].sum().idxmax(),
        df.groupby("Region")["Revenue"].sum().idxmax(),
        df.groupby("Channel")["Revenue"].sum().idxmax()
    ]
})

# --- Append Dashboard to existing Excel ---
with pd.ExcelWriter(excel_path, engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    dashboard.to_excel(writer, sheet_name="Dashboard", index=False)

print(f"✅ Dashboard sheet added at: {excel_path}")
```
